```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5,50,100)
y = np.linspace(10,30,100)
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show
```

```python
y1 = x**2
```

```python
plt.plot(x,y1)
plt.show
```

```python
plt.plot(x,y1,color='red')
plt.show
```

```python
plt.plot(x,y1,color='blue')
plt.show
```

```python
plt.plot(x,y1,color='orang')
plt.show
```

```python
plt.plot(x,y1,color='green')
plt.show
```

```python
plt.plot(x,y1,color='yellow')
plt.show
```

```python
plt.plot(x,y1,color='orange',linewidth = 5,linestyle = '-')
plt.show
```

```python
y2 = np.log(x)
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x,y2)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.show
```
